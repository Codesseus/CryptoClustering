```python
# Import required libraries and dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(market_data_df)
```

```python
# Create a DataFrame with the scaled data
scaled_market_data_df = pd.DataFrame(market_data_scaled, columns=market_data_df.columns)

# Copy the crypto names from the original data
scaled_market_data_df["coin_id"] = market_data_df.index

# Set the coinid column as index
scaled_market_data_df = scaled_market_data_df.set_index("coin_id")

# Display sample data
scaled_market_data_df.head()
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_list = range(1, 11)

# Create an empty list to store the inertia values
inertia_list = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
for k in k_list:
    model = KMeans(n_clusters=k, n_init="auto", random_state=0)
    model.fit(scaled_market_data_df)
    inertia_list.append(model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k":k_list, "inertia":inertia_list}

# Create a DataFrame with the data to plot the Elbow curve
market_data_elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
display(market_data_elbow_df)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
market_data_elbow_df.plot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k_list
)
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:** 4 is the best value for k, because it is where the curve of the graph stabilizes for the rest of its run

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
model_k4 = KMeans(n_clusters=4, n_init="auto", random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model_k4.fit(scaled_market_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_cluster_predictions = model_k4.predict(scaled_market_data_df)

# View the resulting array of cluster values.
print(f"Predicted Crypto Clusters: {crypto_cluster_predictions}")
```

```python
# Create a copy of the DataFrame
prediction_scaled_market_data_df = scaled_market_data_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
prediction_scaled_market_data_df["crypto_cluster"] = crypto_cluster_predictions

# Display sample data
prediction_scaled_market_data_df.head()
```

```python
# Create a scatter plot using Pandas plot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Use "rainbow" for the color to better visualize the data.
prediction_scaled_market_data_df.plot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="crypto_cluster",
    colormap = "rainbow"
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_3 = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
scaled_market_data_pca_array = pca_3.fit_transform(scaled_market_data_df)

# View the first five rows of the DataFrame. 
scaled_market_data_pca_array[:5]
```

```python
# Retrieve the explained variance to determine how much information  can be attributed to each principal component.
print(f"Explained Varience Ratio: {pca_3.explained_variance_ratio_}")
print(f"Total Explained Variance: {sum(pca_3.explained_variance_ratio_)}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:**  0.895 aka 89.5% of the data is captured in PCA's 1, 2, and 3

```python
# Create a new DataFrame with the PCA data.
# Note: The code for this step is provided for you

# Creating a DataFrame with the PCA data
pca_market_data_df = pd.DataFrame(
    scaled_market_data_pca_array, 
    columns=["PCA1", "PCA2", "PCA3"]
)


# Copy the crypto names from the original data
pca_market_data_df["coin_id"] = scaled_market_data_df.index

# Set the coinid column as index
pca_market_data_df = pca_market_data_df.set_index("coin_id")

# Display sample data
pca_market_data_df.head(10)
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_list_pca = range(1, 11)

# Create an empty list to store the inertia values
inertia_list_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using PCA DataFrame.
# 3. Append the model.inertia_ to the inertia list
for k in k_list_pca:
    model_pca = KMeans(n_clusters=k, n_init="auto", random_state=0)
    model_pca.fit(pca_market_data_df)
    inertia_list_pca.append(model_pca.inertia_)

# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k":k_list_pca, "inertia":inertia_list_pca}

# Create a DataFrame with the data to plot the Elbow curve
pca_market_data_elbow_df = pd.DataFrame(pca_elbow_data)

# Display the DataFrame
display(pca_market_data_elbow_df)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_market_data_elbow_df.plot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve PCA", 
    xticks=k_list
)
```

#### Answer the following questions: 
* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**  6 is the best value for k, for the same reason as above, it is where the tangent line of the curve largely stops changing


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Yes, while 4 is where the largest flattening of the tangent line occurs, the curve does not stop changing until k=6

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_k6 = KMeans(n_clusters=6, n_init="auto", random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model_k6.fit(pca_market_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_crypto_clusters_prediction_array = model_k6.predict(pca_market_data_df)

# Print the resulting array of cluster values.
print(f"Predicted Crypto Clusters: {pca_crypto_clusters_prediction_array}")
```

```python
# Create a copy of the DataFrame with the PCA data
predicted_pca_market_data_df = pca_market_data_df.copy()

# Add a new column to the DataFrame with the predicted clusters
predicted_pca_market_data_df["crypto_cluster"] = pca_crypto_clusters_prediction_array

# Display sample data
predicted_pca_market_data_df.head()
```

```python
# Create a scatter plot using hvPlot by setting `x="PCA1"` and `y="PCA2"`. 
predicted_pca_market_data_df.plot.scatter(
    x="PCA1",
    y="PCA2",
    c="crypto_cluster",
    colormap = "rainbow"
)
```

### Determine the Weights of Each Feature on each Principal Component

```python
# Use the columns from the original scaled DataFrame as the index.
pca_market_data_component_weights = pd.DataFrame(
    pca_3.components_.T, 
    columns = pca_market_data_df.columns,
    index = scaled_market_data_df.columns
)

display(pca_market_data_component_weights)
```
